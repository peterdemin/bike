```python
def gaussian(x, mu, sig):
    return math.exp(-(x - mu)*(x - mu) / (2.0 * sig * sig))


def hour_distance(a, b):
    return float(a == b)
#     g = gaussian(a, b, 0.3)
#     if abs(g) > 0.001:
#         return g
#     else:
#         return 0.0
#     return math.exp(abs(a - b) - 1)


def month_distance(a, b):
    return gaussian(a, b, 2)
#     return (a - b) ** 2


[hour_distance(x, 12.0) for x in range(0, 24)]
```

```python
def read_csv(filename):
    with open(filename) as fp:
        reader = csv.reader(fp, delimiter=',')
        reader.next()  # skip header
        for row in reader:
            yield row


def read_train():
    return read_csv('train.csv')


def read_test():
    return read_csv('test.csv')


def parse_date(date_str):
    return datetime.datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')


def make_feature(row, test_set=False):
    date = parse_date(row[0])
    years_ttl = date.year - 2010
    months_ttl = years_ttl * 12 + date.month
    days_ttl = (date - datetime.datetime(2010, 1, 1)).days
    holiday, workingday = bool(row[2]), bool(row[3])
    weekend = int((not workingday) and (not holiday))
    hour_features = tuple([hour_distance(date.hour, h)
                           for h in range(8, 21)])
    date_features = (date.day, years_ttl, months_ttl, weekend, days_ttl, date.hour, month_distance(date.month, 7))
    if test_set:
        input_data = date_features + hour_features + tuple(map(float, row[1:]))
        return input_data
    else:
        input_data = date_features + hour_features + tuple(map(float, row[1:-3]))
        return input_data, int(row[-1])


feature_names = (
    'day',
    'years_ttl',
    'months_ttl',
    'weekend',
    'days_ttl',
    'hour',
    'month_july',
    'hour8',
    'hour9',
    'hour10',
    'hour11',
    'hour12',
    'hour13',
    'hour14',
    'hour15',
    'hour16',
    'hour17',
    'hour18',
    'hour19',
    'hour20',
    'season',
    'holiday',
    'workingday',
    'weather',
    'temp',
    'atemp',
    'humidity',
    'windspeed',
)


def make_key(row):
    return row[0]
    
    
def write_prediction(filename, keys, values):
    with open(filename, 'w') as fp:
        writer = csv.writer(fp, delimiter=',')
        writer.writerow(('datetime', 'count'))
        for row in itertools.izip(keys, values):
            writer.writerow(row)

            
def predict_trimmed(machine, X, left=0, right=10000):
    return [min(right, max(left, y))
            for y in machine.predict(X)]


print make_feature(sample)
print make_feature(sample, True)
```

```python
from sklearn.cross_validation import train_test_split

def split_by_date(xs, ys, day_of_month):
    a, b = [[], []], [[], []]
    for x, y in itertools.izip(xs, ys):
        to = a
        if x[0] > day_of_month:
            to = b
        to[0].append(x)
        to[1].append(y)
    return map(np.array, [a[0], b[0], a[1], b[1]])
        
Xa, Ya = map(np.array, zip(*map(make_feature, read_train())))
X, Xt, Y, Yt = train_test_split(Xa, Ya, test_size=0.2)
X, Xt, Y, Yt = split_by_date(Xa, Ya, day_of_month=15)
Xo = np.array(map(functools.partial(make_feature, test_set=True), read_test()))
Ko = map(np.array, map(make_key, read_test()))

print 'total known', Xa.shape, Ya.shape
print 'training   ', X.shape, Y.shape
print 'testing    ', Xt.shape, Yt.shape
print 'submission ', Xo.shape
```

```python
from sklearn.linear_model import LinearRegression, RidgeCV
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn import grid_search

parameters = {
#     'PCA__n_components': range(5, 20),
    'linear__normalize': (True, False),
    'poly__degree': (1, 2, 3),
    'poly__interaction_only': (True, False)
}
parameters = {
#     'PCA__n_components': [10],
    'linear__normalize': [True],
    'poly__degree': [3],
    'poly__interaction_only': (True, False)
}
pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=2, interaction_only=True)),
#     ('PCA', PCA()),
    ('linear', RidgeCV(alphas=(0.001, 0.01, 0.0001, 0.05, 0.1, 0.5, 1.0), normalize=True, fit_intercept=True)),
])


def best_from_grid():
    grid = grid_search.GridSearchCV(
        estimator=pipeline,
        param_grid=parameters,
        cv=3,
        n_jobs=3,
    )
    grid.fit(Xa, Ya)
    return grid.best_estimator_


def fit_pipeline():
    pipeline.fit(X, Y)
    return pipeline

    
machine = fit_pipeline()
print("Machine parameters:")
best_parameters = machine.get_params()
for param_name in sorted(best_parameters.keys()):
    print("\t%s: %r" % (param_name, best_parameters[param_name]))

Y_ = predict_trimmed(machine, X)
Yt_ = predict_trimmed(machine, Xt)
Yo = predict_trimmed(machine, Xo)
write_prediction('predict.csv', Ko, Yo)
```

```python
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from tabulate import tabulate


def kaggle_score(Y, Y_):
    return (sum([(math.log(y_ + 1) - math.log(y + 1)) ** 2
                  for y, y_ in zip(Y, Y_)]) / len(Y)) ** 0.5


def scores(Y, Yp):
    return (
        explained_variance_score(Y, Yp),
        r2_score(Y, Yp),
        kaggle_score(Y, Yp),
        mean_absolute_error(Y, Yp),
        mean_squared_error(Y, Yp),
    )
headers = ('explained_variance', 'r2', 'kaggle', 'mean_absolute', 'mean_squared')

print tabulate([('train',) + scores(Y, Y_), ('test',) + scores(Yt, Yt_)], headers=headers)
```
